```python
from Battleship_puzzle import Pieza, Barco, Submarino, Tablero, Direccion
```

```python
piezas = {
            "Circle": Pieza.C,
            "Water": Pieza.W,
            "Left": Pieza.L,
            "Right": Pieza.R,
            "Middle": Pieza.M,
            "Top": Pieza.T,
            "Bottom": Pieza.B
        }

direccion_dict = {
    "S": Direccion.VERTICAL,
    "E": Direccion.HORIZONTAL
}

def definir_tablero(instrucciones: str):
    comandos = instrucciones.split("\n")
    pistas = []
    flota = {}
    solucion = []

    for comando in comandos:
        if "START-NEW-BOARD" in comando:
            nFilas, nCols = tuple(map(int, comando[22:].split(" ")))
        
        elif "Hint:" in comando:
            fila, columna, tipo, _ = comando[6:].split(" ")
            pistas.append((int(fila), int(columna), piezas[tipo]))
        
        elif "Row tallies:" in comando:
            cuentasFilas = tuple(map(int,comando[18:].split("  ")))

        elif "Column tallies:" in comando:
            cuentasColumnas = tuple(map(int,comando[18:].split("  ")))
        
        elif "PLACESHIP" in comando:
            tamano, fila_obj, col_obj, direccion = comando[14:].split(" ")
            tamano = int(tamano)
            # guarda tama√±o de flota
            if not flota.get(tamano):
                flota[tamano] = 1
            else:
                flota[tamano] += 1
            
            solucion.append((int(fila_obj), int(col_obj), tamano, direccion_dict[direccion]))


        else:
            continue

    return (nFilas, nCols, flota, pistas, cuentasFilas, cuentasColumnas, solucion)

```
